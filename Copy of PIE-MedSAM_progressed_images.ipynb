```python
#running this:
import os
import shutil

RUN_PIE = "run_pie.py"

images_dir = "/content/drive/MyDrive/genai/GenAIProject/PIE/medsam_finaloutput_test_autodetect_boundingbox_2/images"
masks_dir = "/content/drive/MyDrive/genai/GenAIProject/PIE/medsam_finaloutput_test_autodetect_boundingbox_2/medsam_masks"

output_root = "/content/drive/MyDrive/genai/GenAIProject/PIE_medsam_autobox"

# Create output root directory
os.makedirs(output_root, exist_ok=True)

for filename in sorted(os.listdir(images_dir)):

  image_path = os.path.join(images_dir, filename)

  mask_path  = os.path.join(masks_dir, filename.replace("_slice", "_medsam"))

  patient_id = filename.split("_")[0]
  base_name = filename.replace(".png", "")

  patient_folder = os.path.join(output_root, patient_id)
  os.makedirs(patient_folder, exist_ok=True)

  outdir = os.path.join(patient_folder, base_name)
  os.makedirs(outdir, exist_ok=True)

  print(f"PIE-MedSAM progression for {filename}")

  cmd = (
      f"python {RUN_PIE} "
      f"--pretrained_model_name_or_path 'CompVis/stable-diffusion-v1-4' "
      f"--finetuned_path '/content/drive/MyDrive/genai/GenAIProject/PIE/checkpoints/unet' "
      f"--image_path '{image_path}' "
      f"--mask_path '{mask_path}' "
      f"--prompt 'A glioblastoma MRI image after successful treatment.' "
      f"--output_dir '{outdir}' "
      f"--step 10 "
      f"--strength 0.5 "
      f"--guidance_scale 7.5 "
      f"--seed 42 "
      f"--resolution 512"
  )

  os.system(cmd)
```

```python
import os

RUN_PIE = "run_pie.py"

image_name = "Patient-008_week-000_CT1_slice.png"

images_dir = "/content/drive/MyDrive/genai/GenAIProject/PIE/data/gbm/finaloutput_test/images/"
masks_dir  = "/content/drive/MyDrive/genai/GenAIProject/PIE/data/gbm/finaloutput_test/masks/"
output_root = "../PIE_results_all_test"

image_path = os.path.join(images_dir, image_name)
mask_path  = os.path.join(masks_dir, image_name.replace("_slice", "_mask"))

patient_id = image_name.split("_")[0]
base_name  = image_name.replace(".png", "")

patient_folder = os.path.join(output_root, patient_id)
os.makedirs(patient_folder, exist_ok=True)

outdir = os.path.join(patient_folder, base_name)
os.makedirs(outdir, exist_ok=True)


cmd = (
    f"python {RUN_PIE} "
    f"--pretrained_model_name_or_path 'CompVis/stable-diffusion-v1-4' "
    f"--finetuned_path '/content/drive/MyDrive/genai/GenAIProject/PIE/checkpoints/unet' "
    f"--image_path '{image_path}' "
    f"--mask_path '{mask_path}' "
    f"--prompt 'A glioblastoma MRI image after successful treatment.' "
    f"--output_dir '{outdir}' "
    f"--step 10 "
    f"--strength 0.5 "
    f"--guidance_scale 7.5 "
    f"--seed 42 "
    f"--resolution 512"
)

os.system(cmd)

```
