```python
def medsam_predict(image_path, save_path):

  img = cv2.imread(image_path)
  rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  predictor.set_image(rgb)


  #full image bounding box:
  h, w = rgb.shape[:2]
  box = np.array([0, 0, w, h])

  masks, scores, logits = predictor.predict(box=box, multimask_output=False)

  mask = masks[0].astype(np.uint8) * 255
  cv2.imwrite(save_path, mask)
```

```python
#using size filtering where min area ratio means too tiny and noise, and max area ratio means its too big and probably wrong

def find_tumor_bbox(rgb, bright_percentile=99, min_area_ratio= 0.001, max_area_ratio= 0.4, margin= 5):
  gray = cv2.cvtColor(rgb, cv2.COLOR_RGB2GRAY)

  #masks
  blank2, brain_mask = cv2.threshold(gray, 5, 255, cv2.THRESH_BINARY)
  coords = cv2.findNonZero(brain_mask)
  if coords is None:
    return None
  x_b, y_b, w_b, h_b = cv2.boundingRect(coords)
  brain_roi = gray[y_b:y_b+h_b, x_b:x_b+w_b]


  #rregions on flair images are bright
  brain_pixels = brain_roi[brain_roi > 0]


  if brain_pixels.size == 0:
    return None


  thr = np.percentile(brain_pixels, bright_percentile)
  bright = (brain_roi >= thr).astype(np.uint8) * 255

  kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
  bright = cv2.morphologyEx(bright, cv2.MORPH_OPEN, kernel, iterations=1)

  bright = cv2.morphologyEx(bright, cv2.MORPH_CLOSE, kernel, iterations=1)

  #largest brightest blob/region

  num_labels, labels, stats, blank = cv2.connectedComponentsWithStats(bright, 8)
  img_area = gray.shape[0] * gray.shape[1]

  candidates = []

  for i in range(1, num_labels):
    area = stats[i, cv2.CC_STAT_AREA]
    if not (min_area_ratio * img_area <= area <= max_area_ratio * img_area):
      continue

    x_i, y_i, w_i, h_i, _ = stats[i]
    candidates.append((area, x_i + x_b, y_i + y_b, w_i, h_i))

  #  print(candidates)

  # if not candidates:
  #   return None


  #largest remaining candidate (or j basically largest bright structure)
  blank3, bx, by, bw, bh = max(candidates, key=lambda c: c[0])

  #create final bounding box::

  x1 = max(0, bx - margin)
  y1 = max(0, by - margin)
  # print(x1)
  # print(y1)



  x2 = min(gray.shape[1] - 1, bx + bw + margin)
  y2 = min(gray.shape[0] - 1, by + bh + margin)
  # print(x2)
  # print(y2)


  return np.array([x1, y1, x2, y2])

```

```python
from PIL import Image
import os

def medsam_predict_auto(image_path, save_path):

  #attempt image reading :

  img_bgr = cv2.imread(image_path)
  if img_bgr is None:
    print("image undreadablee", image_path)
    return False

  rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

  predictor.set_image(rgb)
  box = find_tumor_bbox(rgb)

  #using standard centered bounding box if we can't find tumor
  if box is None:

    h, w = rgb.shape[:2]

    cx1 = int(w*0.2)
    cx2 = int(w*0.8)
    cy1 = int(h*0.2)
    cy2 = int(h*0.8)

    # print(cx1)
    # print(cx2)
    # print(cx3)
    # print(cx4)

    box = np.array([cx1, cy1, cx2, cy2])


    print("centered box is used for:", os.path.basename(image_path))

  #running medsam
  masks, scores, logits = predictor.predict(box=box, multimask_output=False)
  mask = (masks[0].astype(np.uint8) * 255)

  mask = 255 - mask

  os.makedirs(os.path.dirname(save_path), exist_ok=True)
  cv2.imwrite(save_path, mask)

  return True
```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm
import shutil

test_images_dir = "/content/drive/MyDrive/genai/GenAIProject/PIE/data/gbm/finaloutput_test/images"
output_root = "/content/drive/MyDrive/genai/GenAIProject/PIE/medsam_finaloutput_test_autodetect_boundingbox"

images_out_dir = os.path.join(output_root, "images")
masks_out_dir  = os.path.join(output_root, "medsam_masks")

os.makedirs(images_out_dir, exist_ok=True)
os.makedirs(masks_out_dir, exist_ok=True)

print("results in:", output_root)

count = 0
for filename in tqdm(sorted(os.listdir(test_images_dir))):

  image_path = os.path.join(test_images_dir, filename)
  mask_save_path = os.path.join(masks_out_dir, filename.replace("_slice", "_medsam"))


  #copy images to images
  img_copy_path = os.path.join(images_out_dir, filename)
  if not os.path.exists(img_copy_path):
   shutil.copy(image_path, img_copy_path)



  #run MedSAM auto bbox mask
  success = medsam_predict_auto(image_path, mask_save_path)

  if success:
    count += 1
  else:
    print("failed: ", filename)

print(f"Medsam mask count: {count}")
print("Masks in ", masks_out_dir)
```
