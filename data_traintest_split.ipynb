```python
import os, shutil, random

input_slice_dir = "data/gbm/finaloutput"
input_mask_dir = "data/gbm/masks"

train_dir_img = "data/gbm/finaloutput_train/images"
train_dir_mask = "data/gbm/finaloutput_train/masks"

test_dir_img = "data/gbm/finaloutput_test/images"
test_dir_mask = "data/gbm/finaloutput_test/masks"

split_ratio = 0.8

os.makedirs(train_dir_img, exist_ok=True)
os.makedirs(train_dir_mask, exist_ok=True)
os.makedirs(test_dir_img, exist_ok=True)
os.makedirs(test_dir_mask, exist_ok=True)

files = sorted(os.listdir(input_slice_dir))

#splitting by patient id:
patient_id_set = set()
for file_name in files:
  parts = file_name.split("_")
  patient_id = parts[0]
  patient_id_set.add(patient_id)
patient_ids = sorted(list(patient_id_set))

random.shuffle(patient_ids)
split_index = int(len(patient_ids) * split_ratio)

train_patients = set(patient_ids[:split_index])
test_patients = set(patient_ids[split_index:])

for filename in files:
  patient = filename.split("_")[0]
  slice_src = os.path.join(input_slice_dir, filename)

  mask_filename = filename.replace("_slice", "_mask")
  mask_src = os.path.join(input_mask_dir, mask_filename)

  if patient in train_patients:
    slice_dest = os.path.join(train_dir_img, filename)
    mask_dest = os.path.join(train_dir_mask, mask_filename)
  else:
    slice_dest = os.path.join(test_dir_img, filename)
    mask_dest = os.path.join(test_dir_mask, mask_filename)

  shutil.copy(slice_src, slice_dest)
  shutil.copy(mask_src, mask_dest)
```
